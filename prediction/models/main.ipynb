```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.20,random_state=43)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
def models(model,trainX,testX,trainY,testY):
    model.fit(trainX.values,trainY)
    predictions=model.predict(testX)
    predictions=pd.DataFrame(predictions)
    scores=model.score(testX, testY)
    
    return scores
```

```python
model1=LinearRegression()

model1.fit(x_train.values,y_train)
predictions=model1.predict(x_test)
predictions=pd.DataFrame(predictions)
score1=model1.score(x_test, y_test)

print(score1)
```

```python
model2=DecisionTreeRegressor()
score2=models(model2,x_train,x_test,y_train,y_test)
print(score2)
```

```python
model3=RandomForestRegressor()
score3=models(model3,x_train,x_test,y_train,y_test)
print(score3)
```
